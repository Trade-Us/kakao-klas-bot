```python
import threading
import logging

def get_logger():
    logger = logging.getLogger("Thread Example")
    logger.setLevel(logging.DEBUG)
    #fh = logging.FileHandler("threading.log")
    fh = logging.StreamHandler()
    fmt = '%(asctime)s - %(threadName)s - %(levelname)s - %(message)s'
    formatter = logging.Formatter(fmt)
    fh.setFormatter(formatter)
    
    logger.addHandler(fh)
    return logger

def execute(number, logger):
    '''
    쓰레드에서 실행 할 함수(로깅 사용)
    '''
    logger.debug('execute function executing')
    result = number * 2
    logger.debug('execute function ended with : {}'.format(result))
    
logger = get_logger()
for i, name in enumerate(['Kim', 'Lee', 'Park', 'Cho', 'Hong']):
    my_thread = threading.Thread(target=execute, name=name, args=(i, logger))
    my_thread.start()
```

## Thread 동기화 문제 

```python
# 동기화가 구현되지 않았다.
import threading

tot = 0

def add_total(amount):
    '''
    쓰레드에서 실행 할 함수 전역변수 tot에 amount더하기
    '''
    global tot
    tot += amount
    print(threading.currentThread().getName() + ' Not Synchronized :', tot)
    
for i in range(10000):
    my_thread = threading.Thread(target=add_total, args=(1,))
    my_thread.start()
```

```python
import threading

tot = 0
lock = threading.Lock()

def add_total(amount):
    global tot
    lock.acquire()
    try:
        tot += amount
        print(threading.currentThread().getName() + ' Synchronized :', tot)
    finally:
        lock.release()
    '''
    or
    global tot
    with lock:
        total += amount
    print(threading.currentThread().getName() + ' Synchronized :', tot)
    with 문으로 심플하게 가능
    '''

for i in range(10000):
    my_thread = threading.Thread(target=add_total, args=(1,))
    my_thread.start()
```

## 데이터 전달

```python
import threading

from queue import Queue

def creator(data, q):
    print('Creating data and putting it on the queue')
    print('\n')
    for item in data:
        evt = threading.Event()
        q.put((item, evt))
        print('Waiting for data to be doubled')
        evt.wait()
        
def consumer(q):
    while True:
        data, evt = q.get()
        print('Receive Original Data : {}'.format(data))
        processed = data * 5
        print('Receive Processed Data : {}'.format(processed))
        print('\n')
        evt.set()
        q.task_done()
        
q = Queue()
data = [7, 14, 39, 59, 77, 1, 109, 99, 167, 920, 1035]
thread_one = threading.Thread(target=creator, args=(data, q))
thread_two = threading.Thread(target=consumer, args=(q, ))
thread_one.start()
thread_two.start()

q.join()
```

## Thread Timer 반복

```python
import time
import threading

def thread_run():
    print('=====', time.ctime(), "=====")
    for i in range(1, 10):
        # 개발 코드
        print('Thread running - ', i)
        
    threading.Timer(2.5, thread_run).start()

thread_run()
```

## Multiprocessing

```python
import time

# 시작시간
start_time = time.time()

# 멀티 쓰레드 사용하지 않은 경우(20만 카운트)
def count(name):
    for i in range(1, 501):
        print(name, " : ", i)

num_list = ['p1', 'p2', 'p3', 'p4']

for num in num_list:
    count(num)

print("--- %s seconds ---"% (time.time() - start_time))
```

```python
import multiprocessing
import time

# 시작시간
start_time = time.time()

# 멀티쓰레드 사용하는 경우 (20만 카운트)
# pool 사용해서 함수 실행을 병렬
def count(name):
    for i in range(1, 501):
        print(name, " : ", i)

num_list = ['p1', 'p2', 'p3', 'p4']

pool = multiprocessing.Pool(processes=2) # 현재 시스템에서 사용 할 프로세스 개수
pool.map(count, num_list)
#리소스 방지를 위한 close 호출 및 작업 완료 대기 위한 join 호출
pool.close()
pool.join()


print("--- %s seconds ---"% (time.time() - start_time))
```
