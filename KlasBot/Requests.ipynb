### GET요청시 GET인자 지정하기
GET 요청시 인자는 dict와 Key,Value형식의 tuple 두가지로 지정할 수 있습니다.
- dict지정: 동일 Key에 다수 지정 불가(Key: 인자=> 1:1)
- tuple지정: 동일 Key에 다수 지정 가능(Key: 인자=> 1:N)


```python
import requests

#dict 지정

get_params = dict([('k1', 'v1'), ('k1', 'v3'), ('k2', 'v2')])

response = requests.get("http://httpbin.org/get", params=get_params)
print(response.json())
```

```python
import requests

# tuple 지정

get_params = (('k1', 'v1'), ('k1', 'v3'), ('k2', 'v2'))

response = requests.get("http://httpbin.org/get", params=get_params)
print(response.json())
```

### POST 요청시 커스텀헤더, GET인자 지정

```python
import requests

request_headers = {
    'User-Agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36'),
    'Referer': 'http://httpbin.org',
}

get_params = {'k1': 'v1', 'k2':'v2'}
response = requests.post(
    'http://httpbin.org/post',
    headers=request_headers,
    params=get_params
)

print(response)
```

그런데, post 요청에도 get_params를 해주는 이유가 뭘까?  

params 인자로는 GET 요청의 params가 옴을 기억하자  
그러나 지금 POST 요청에도 가능하니, POST는 GET인자를 사용할 수 있는 것이다.
- GET인자: 웹 주소에서 ?Key=Value?.. 형식

### 일반적인 form 전송/요청

- dict 지정 : 같은 키 중복 불가
- tuple 지정 : 같은 키 중복 가능

- BeautifulSoup4 사용하기
- web사이트에 접속이 안되는 경우, 접근 차단 된 것으로, 헤더를 입력하면 괜찮다.

### find를 통해 태그 하나씩 찾기

```python
import requests
from bs4 import BeautifulSoup

request_headers = {
    'User-Agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36'),
    'Referer': 'https://www.melon.com/chart/index.htm', 
}

html = requests.get('https://www.melon.com/chart/index.htm', headers=request_headers,).text
soup = BeautifulSoup(html, 'html.parser')

tag_list = []
for tr_tag in soup.find(class_='service_list_song').find_all('tr'):
    tag = tr_tag.find(class_='wrap_song_info')
    if tag:
        tag_sub_list = tag.find_all(href=lambda value: (value and 'playSong' in value))
        tag_list.extend(tag_sub_list)

for idx, tag in enumerate(tag_list, 1):
    print(idx, tag.text)
```

### 태그관계를 지정해서 찾기(CSS Selector 사용)

CSS Selector를 통한 Tag찾기 지원
- tag name: "tag_name"
- tag id: "#tag_id"  (#을 사용하면 id)
- tag class names: ".tag_class"  (.을 사용하면 class)

#### CSS Selector Syntax(문법)
- *: 모든 Tag
- tag: 해당 모든 Tag
- Tag1 > Tag2: Tag1의 직계인 모든 Tag2


- tag[attr]: attr속성이 정의된 모든 Tag
- tag[attr="bar"]: attr속성이 "bar"문자열과 일치하는 모든 Tag
- tag[attr*="bar"]: attr속성이 "bar"문자열과 부분 매칭되는 모든 Tag
- tag[attr^="bar"]: attr속성이 "bar"문자열로 시작되는 모든 Tag
- tag[attr$="bar"]: attr속성이 "bar"문자열로 끝나는 모든 Tag



```python
import requests
from bs4 import BeautifulSoup

request_headers = {
    'User-Agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36'),
    'Referer': 'https://www.melon.com/chart/index.htm', 
}

html = requests.get('https://www.melon.com/chart/index.htm', headers=request_headers).text
soup = BeautifulSoup(html, 'html.parser')

tag_list = soup.select('.service_list_song tr .wrap_song_info a[href*=playSong]')

for idx, tag in enumerate(tag_list):
    print(idx, tag.text)
```
